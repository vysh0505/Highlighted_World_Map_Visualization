```python
import geopandas as gpd
import matplotlib.pyplot as plt
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
countries_of_interest = ['India', 'Finland', 'Egypt', 'Mexico', 'Libya', 'Canada', 'Malaysia', 'Colombia', 'Angola', 'North korea']
selected_countries = world[world['name'].isin(countries_of_interest)]
world.plot(figsize=(10, 6))
selected_countries.plot(ax=plt.gca(), color='red', marker='o', markersize=50)
plt.title('World Map with Selected Countries')
plt.show()
```
