```python
import geopandas as gpd
import matplotlib.pyplot as plt
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
countries_of_interest = ['India', 'Finland', 'Egypt', 'Mexico', 'Algeria', 'Canada', 'Malaysia', 'Colombia', 'Angola', 'Iceland']
selected_countries = world[world['name'].isin(countries_of_interest)]
fig, ax = plt.subplots(figsize=(10, 6))
world.plot(ax=ax)
selected_countries.plot(ax=ax, color='red', marker='o', markersize=50)

# Add country names as annotations
for idx, row in selected_countries.iterrows():
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, row['name'], fontsize=8, ha='center')

plt.title('NUMBER OF DATA ANALYST ON THE WORLD')
plt.show()
```
